```python
## Implementing the ADX calculator
UpMove = today's high − yesterday's high
DownMove = yesterday's low − today's low
if UpMove > DownMove and UpMove > 0, then +DM = UpMove, else +DM = 0
if DownMove > UpMove and DownMove > 0, then −DM = DownMove, else −DM = 0
```

```python
import csv
import pandas as pd
import urllib
import urllib2
import datetime
import numpy as np
```

```python
dataDict = {};
times = [];
closing = [];
high = [];
low = [];

with open('../Data/getprices_AMD_long.txt', 'rb') as csvfile:
    pricereader = csv.reader(csvfile, delimiter=' ', quotechar='|')
    for i, row in enumerate(pricereader):
        if i > 6:
            dataDict[i] = row;
            values = row[0].split(',');
            times.append(values[0]);
            closing.append(float(values[1]));
            high.append(float(values[2]));
            low.append(float(values[3]));
```

```python
timesUTC = [];

currUnixTime = 0;
currTime = None;

for i in range(len(times)):
    if times[i][0] == 'a':
        #print i;
        currUnixTime = int(times[i][1:]);
        #print currUnixTime;
        currTime = datetime.datetime.utcfromtimestamp(currUnixTime)
        #print currTime
        timesUTC.append(currTime);
    else:
        #print currUnixTime
        timesUTC.append(datetime.datetime.utcfromtimestamp(currUnixTime + int(times[i])));
```

```python
UpMove = [high[n] - high[n-1] for n in range(1,len(high))];
DownMove = [low[n] - low[n-1] for n in range(1, len(low))];
```

```python
positive_DM = [];
negative_DM = [];

for i in range(len(UpMove)):
    if UpMove[i] > DownMove[i] and UpMove[i] > 0:
        positive_DM.append(UpMove[i]);
    else:
        positive_DM.append(0);
    
    if DownMove[i] > UpMove[i] and DownMove[i] > 0:
        negative_DM.append(DownMove[i]);
    else:
        negative_DM.append(0);
```

```python
print negative_DM
```

```python
TR = [];

for i in range(len(high)):
    TR.append(max(high[i] - low[i], abs(high[i] - low[i - 1]), abs(low[i] - closing[i - 1])));
```

```python
def calculateDM_TR_range(positive_DM, negative_DM, TR, n):
    currentDM_positive = 0;
    currentDM_negative = 0;
    currentTR = 0;
    
    DM_n_positive_storage = [];
    DM_n_negative_storage = [];
    TR_storage = [];
    
    DM_n_positive_means = [];
    DM_n_negative_means = [];
    TR_means = [];
    
    for i in range(n):
        # Append all values to DM_n_positive and DM_n_negative;
        DM_n_positive_storage.append(positive_DM[i]);
        DM_n_negative_storage.append(negative_DM[i]);
        TR_storage.append(TR[i]);
        
    currentDM_positive = np.sum(DM_n_positive_storage);
    currentDM_negative = np.sum(DM_n_negative_storage);
    currentTR = np.sum(TR_storage);
    
    DM_n_positive_means.append(currentDM_positive);
    DM_n_negative_means.append(currentDM_negative);
    TR_means.append(currentTR);
    
    for i in range(n, len(positive_DM)):
        # print i;
        del DM_n_positive_storage[0]; # Delete first index
        del DM_n_negative_storage[0];
        del TR_storage[0];
        DM_n_positive_storage.append(positive_DM[i]); # Add data at i;
        DM_n_negative_storage.append(negative_DM[i]);
        TR_storage.append(TR[i]);
        currentDM_positive = np.sum(DM_n_positive_storage);
        currentDM_negative = np.sum(DM_n_negative_storage);
        currentTR = np.sum(TR_storage);
        DM_n_positive_means.append(currentDM_positive);
        DM_n_negative_means.append(currentDM_negative);
        TR_means.append(currentTR);
        
    return DM_n_positive_means, DM_n_negative_means, TR_means;
    
```

```python
DM_14_positive, DM_14_negative, TR_14_means = calculateDM_TR_range(positive_DM, negative_DM, TR, 14);
```

```python
DM_14_positive_np = np.asarray(DM_14_positive);
DM_14_negative_np = np.asarray(DM_14_negative);
TR_means_np = np.asarray(TR_14_means);
```

```python
DI_14_positive = np.multiply(100, np.divide(DM_14_positive_np, TR_means_np));
```

```python
DI_14_positive
```

```python
DI_14_negative = np.multiply(100, np.divide(DM_14_negative_np, TR_means_np));
```

```python
DI_14_negative
```

```python
DI_14_diff = np.abs(np.subtract(DI_14_positive, DI_14_negative));
DI_14_sum  = np.add(DI_14_positive, DI_14_negative);
```

```python
DX = np.multiply(100,  np.divide(DI_14_diff,  DI_14_sum));
```

```python
def calculateADX(DX, n):
    currentADX = 0;
    internalValues = [];
    ADX = [];

    for i in range(n):
        # Append all values to internalValues;
        # print i;
        internalValues.append(DX[i]);
        
    currentADX = np.mean(internalValues);
    ADX.append(currentADX);
        
    for i in range(n, len(DX)):
        # print i;
        del internalValues[0]; # Delete first index
        internalValues.append(DX[i]); # Add data at i;
        currentADX = np.mean(internalValues);
        ADX.append(currentADX);
        
    return ADX;
```

```python
ADX = calculateADX(DX, 14);
```
